```python
import pandas as pd 
from pathlib import Path
```

```python
file = ('Resources/titanic.csv')
```

```python
df = pd.read_csv(file)
```

```python
df.head()
```

```python
titanic_df = df[['Name', 'Sex', 'Age', 'Pclass', 'Survived']]
titanic_df.head()
```

```python
titanic_df.set_index('Name')
titanic_df.head()
```

```python
titanic_df = titanic_df.rename(columns={'Pclass': 'Ticket Class'})
titanic_df.head()
```

```python
titanic_df['Survived'] = titanic_df['Survived'].replace([0], 'No')
titanic_df['Survived'] = titanic_df['Survived'].replace([1], 'Yes')
titanic_df.head()
```

```python
titanic_survivor_df = titanic_df.loc[(titanic_df['Survived'] == 'Yes')]
titanic_survivor_df.head()
```

```python
titanic_df.to_csv(index=False)
```
